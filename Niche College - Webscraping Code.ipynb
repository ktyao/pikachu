```python
school_name = []
ranking = []
location = []
acceptance_rate = []
```

```python
# Specify url
# Read url from website
the_rankings = "https://www.timeshighereducation.com/world-university-rankings/2018/world-ranking#!/page/0/length/25/subjects/3066/sort_by/rank/sort_order/asc/cols/stats"
qs_rankings = "https://www.university-list.net/rank/univ-150038.html"
top_rankings = "https://www.topuniversities.com/university-rankings/university-subject-rankings/2018/engineering-technology"
```

```python
# Loop through Niche College website to scrape

for i in range(20):
    niche_rankings = "https://www.niche.com/colleges/search/best-colleges-for-engineering/?page="+str(i+1)
    page = requests.get(niche_rankings)
    page = page.text
# query the website and return the html to the variable ‘page’
#page = urllib.request.urlopen(qs_rankings)

#file = open("table_data.csv","w",newline='')
#writer = csv.writer(file)

# Use page to create a BeautifulSoup object and specify the type of parser
    soup = bs(page, 'html.parser')
#soup = bs(page,"lxml")

    school_name += [s.text for s in soup.select("li > div > div > a > div > h2")]
    ranking += [s.text for s in soup.select("li > div > div > a > div > div > span")]
    location += [s.text for s in soup.select("li > div > div > a > div > ul > li")]
    acceptance_rate += [s.text for s in soup.select("li > div > div > a > div > ul > li > div > span")]
```

## Start cleaning raw data

#### Raw data must be saved to csv and then cleaned up for further use

```python
# Save raw data to csv
raw = []
raw = school_name + location + ranking + acceptance_rate
raw_data = pd.DataFrame(raw)
raw_data.to_csv("raw_data_Niche_2017.csv", sep=',', encoding='utf-8')
```

```python
a = [x for x in acceptance_rate if x]  # Remove empty elements
```

```python
# Find the location of each school
locat=[]
for s in location:
    if ' ' in s and 'Year' not in s and 'Overall' not in s and 'SAT' not in s:
        if 'Acceptance' not in s and 'Net' not in s:
            locat.append(s)
locat.insert(404,'Albany, NY')    # Albany was missing... this is to insert the missing value at the specific index
```

```python
# Store relevant characteristics

net_price = acceptance_rate[4::8]
accept_rate = acceptance_rate[2::8]
sat_range = acceptance_rate[6::8]
```

```python
# Create pandas dataframe

Schools = pd.DataFrame({"Price": net_price, "Acceptance Rate": accept_rate, "SAT Range": sat_range, "School Name": school_name, "Location": locat})
Schools
```
