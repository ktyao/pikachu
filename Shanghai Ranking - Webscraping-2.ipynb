```python
# Webscrape Shanghai Rankings
# Subject Rankings

import os
import re
import sys
import csv
from bs4 import BeautifulSoup as bs
import requests, matplotlib.pyplot
import numpy as np
import pandas as pd
from urllib.request import urlopen
```

```python
# Specify url and open website
url = "http://www.shanghairanking.com/ARWU2017.html"
page = requests.get(url)
page = page.text
```

## Start parsing raw data. Storing and cleaning after

```python
# Use page to create a BeautifulSoup object and specify the type of parser
# Raw data
soup = bs(page, 'html.parser')
#soup = bs(page,"lxml")
```

```python
# Parse raw_data
raw_data = soup.select('div')
raw_data = [s.text for s in soup.select('div')]
```

```python
# raw_data dataframe
Raw = pd.DataFrame(raw_data)
```

```python
# Raw data to csv
Raw.to_csv("raw_data_Shanghai_2017.csv", sep=',', encoding='utf-8')
```

```python
# Cleaning raw data

header = [s.text for s in soup.select("tr > th")]
loc = soup.select('td > a > img')  # Finding location
school_name = [s.text for s in soup.select("tr > td > a")]  #Raw school names
school_name = school_name[0::2]

# Raw scores, ranking, total score and award staff
score = [s.text for s in soup.select("tr > td > div")]
nat_ranking = score[0::8] # National Ranking
total_score = score[1::8] # Overall Ranking
award_staff = score[3::8] # Staff awarded Nobel Prize while at institution
# score variable 5 scores per school
# First: Total Score
# Second: Highly Cited Researchers
# Third: Papers indexed in Science citations
# Fourth: Papers published in Top 20% Journals
# Fifth: Expenditures / Funding
```

```python
# Create pandas dataframe
Schools = pd.DataFrame({"Total Score": total_score, "National Ranking": nat_ranking, "Staff Awarded Nobel Prize": award_staff, "School Name": school_name, "Location": loc})
Schools
```
